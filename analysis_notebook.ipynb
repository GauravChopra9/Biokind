## Find the state/province that donated the most in total

```python
df_2020_cleaned_state = df_2020.copy()
df_2020_cleaned_state.fillna('Unknown', inplace=True)
df_2020_cleaned_state = df_2020_cleaned_state.applymap(lambda x: x.strip() if isinstance(x, str) else x) # strip everything
df_2020_cleaned_state.head()
```

```python
df_2020_cleaned_state.groupby('State/Province').agg({'Donation Amount': 'sum'}).reset_index().sort_values(by='Donation Amount', ascending=False)
# Wisconsin donated in total 38080.33, a lot more than other states
# Update 10/31: lots of Unknown donations wtf
```

```python
df_2020_cleaned_state.groupby('State/Province').agg({'Donation ID': 'count'}).reset_index().sort_values(by='Donation ID', ascending=False)
# Wisconsin donated 303 times, a hell lot more than others, with Minnesota coming in 2nd at 12
# Update 10/31: lots of Unknown donations
```

```python
df_2020_cleaned_state.groupby('Account Type').agg({'Donation Amount': 'sum'}).reset_index()
# Companies donated much less in total
```

```python
df_2020_cleaned_state.groupby('Account Type').agg({'Donation ID': 'count'}).reset_index()
# Only 10 companies donated lmao
# Update: 30 companies donated, I forgot to account for unknown guys
```

```python
df_2020_cleaned_state[df_2020_cleaned_state['Account Type'] == 'Company']
# At first glance, most companies donations are Unknown
```

```python
df_2020_cleaned_state_wionly = df_2020_cleaned_state[df_2020_cleaned_state['State/Province'] == 'WI']
df_2020_cleaned_state_wionly['City'] = df_2020_cleaned_state_wionly['City'].astype(str)
df_2020_cleaned_state_wionly['City'] = df_2020_cleaned_state_wionly['City'].apply(lambda x : x.lower()) # make city names all lowercase
df_2020_cleaned_state_wionly['City'] = df_2020_cleaned_state_wionly['City'].apply(lambda x : x.strip()) # apply strip
```

```python
df_2020_cleaned_state_wionly.groupby('City').agg({'Donation Amount': 'sum'}).reset_index().sort_values(by='Donation Amount', ascending=False)
# Madison city donated the most
```

```python
df_2020_cleaned_state_wionly[df_2020_cleaned_state_wionly['City'] == 'unknown'] # Highlighting the unknown donations of WI, not much
```

```python
df_2020_cleaned_state_wionly.groupby('City').agg({'Donation ID': 'count'}).reset_index().sort_values(by='Donation ID', ascending=False)
# Madison donated the most
# I would like to map these on a heatmap, but idk how to do that
```

# Compare 2020 and 2021 side by side

```python
df_2020_cleaned_state['Year'] = 2020 # Add year number to help with combining
df_2021_cleaned_state['Year'] = 2021
df_2020_cleaned_state.head()
```

```python
df_combined = pd.concat([df_2020_cleaned_state, df_2021_cleaned_state])
df_combined
```

```python
df_combined_grouped = df_combined.groupby(['State/Province', 'Year']).agg({'Donation Amount': 'sum', 'Donation ID': 'count'}).reset_index().sort_values(by='Year')
```

```python
df_combined_grouped['Donation Amount in 2020'] = df_combined_grouped['Donation Amount'] * (df_combined_grouped['Year'] == 2020)
df_combined_grouped['Donation Count in 2020'] = df_combined_grouped['Donation ID'] * (df_combined_grouped['Year'] == 2020)
df_combined_grouped['Donation Amount in 2021'] = df_combined_grouped['Donation Amount'] * (df_combined_grouped['Year'] == 2021)
df_combined_grouped['Donation Count in 2021'] = df_combined_grouped['Donation ID'] * (df_combined_grouped['Year'] == 2021)
df_combined_grouped.drop(axis=1, columns=['Donation ID'], inplace=True)

df_combined_grouped
```

```python
donation_by_state_2020 = df_combined_grouped.groupby('State/Province').agg({'Donation Amount in 2020': 'sum'}).reset_index().sort_values(by='Donation Amount in 2020', ascending=False)
donation_by_state_2020 = donation_by_state_2020[donation_by_state_2020['Donation Amount in 2020'] > 0]
donation_by_state_2020
```

```python
donation_by_state_2020[donation_by_state_2020['State/Province'] != 'Unknown'].to_csv('intermediate_csvs/donation_by_state_2020.csv', index=False)
# Export the file, will import to Tableau bcuz Geopandas refused to work
```

```python
donation_by_state_2021 = df_combined_grouped.groupby('State/Province').agg({'Donation Amount in 2021': 'sum'}).reset_index().sort_values(by='Donation Amount in 2021', ascending=False)
donation_by_state_2021 = donation_by_state_2021[donation_by_state_2021['Donation Amount in 2021'] > 0]
donation_by_state_2021
```

```python
donation_by_state_2021[donation_by_state_2021['State/Province'] != 'Unknown'].to_csv('intermediate_csvs/donation_by_state_2021.csv', index=False)
```
